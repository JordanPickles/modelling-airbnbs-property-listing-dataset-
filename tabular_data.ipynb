```python
import pandas as pd
import ast
from ast import literal_eval
```

```python
df = pd.read_csv('./airbnb-property-listings/tabular_data/listing.csv')
df.info()
```

```python
df = df.dropna(subset = ["Cleanliness_rating", "Accuracy_rating", "Communication_rating", "Location_rating", "Check-in_rating", "Value_rating"])
df
```

```python
df.dropna(subset = ['Description'])
df['Description'] = df['Description'].astype(str)
df['Description'] = df['Description'].str.replace('About this space', '')
df['Description'] = df['Description'].str.replace('"', '')
df['Description'] = df['Description'].str.split(",")
df['Description'] = df['Description'].apply(lambda x: [i for i in x if i != ''])
df['Description'] = df['Description'].apply(lambda x: ''.join(x))
print(df['Description'][2])
```

```python
df['Description'] = df['Description'].apply(lambda x: ast.literal_eval(x))

# Remove empty quotes and "About this space" prefix from the lists
df['Description'] = df['Description'].apply(lambda x: [i.replace('"', '').replace('About this space', '') for i in x if i != ''])

# Join the list items with a whitespace
df['Description'] = df['Description'].apply(lambda x: ' '.join(x))

# Remove records with a missing description
df = df[df['Description'].notnull()]

```

```python
df['Description'] = df['Description'].str.replace('About this space', '')
df['Description'] = df['Description'].apply(lambda x: [item for item in eval(str(x)) if item != ''])
df['Description'] = df['Description'].apply(lambda x: ''.join(x))
```
